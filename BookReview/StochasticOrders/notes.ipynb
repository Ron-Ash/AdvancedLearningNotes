Comparison between two distribution functions us commonly done via their associated means; and when they are equal (or centred about the same value), then their standard deviations are instead investigated. These however, are not very informative as they reduce the entire distributions into 2 pairs of numbers, and in many cases are undefined. *Stochastic Orders* provide a more robust framework for comparing statistical distributions by capturing differences in their shapes, spreads, tail behaviours, and other structural properties.

# Univariate Stochastic Orders
Let $X$ and $Y$ be two random variables such that $$\mathbb{P}(X>x)\leq\mathbb{P}(Y>x)$$ for all $x\in\mathbb{R}$ (same as $\mathbb{P}(X\leq x)\geq\mathbb{P}(Y\leq x)$). $X$ is said to be *smaller than* $Y$ (in the usual *stochastic order*), denoted as $X\leq_{st}Y$; and can be interpreted as "$X$ beign less likely to take on large values than $Y$". Additionally note that $X\leq_{st}Y\rightarrow\int_{x}^{\infty}\mathbb{P}(Y>y)-\mathbb{P}(X>y)dy$ is decreasing in $x\in\mathbb{R}$.

These can be rewritten as $\mathbb{E}[\mathbb{I}_U(X)]\leq\mathbb{E}[\mathbb{I}_U(Y)]$ for all *upper sets* (set $S\subseteq X$ such that $\forall s\in S,\forall x\in X, s<x\rightarrow x\in S$) $U\subseteq(-\infty,\infty)$ where $\mathbb{I}_U$ denotes the indicator function of $U$. And from this it follows that if $X\leq_{st}Y$, then 
$$\mathbb{E}[\sum_{i=1}^ma_i\mathbb{I}_{U_i}(X)]-b\leq \mathbb{E}[\sum_{i=1}^ma_i\mathbb{I}_{U_i}(Y)]-b\overset{\lim_{m\rightarrow\infty}}{\rightarrow}\mathbb{E}[\phi(X)]\leq \mathbb{E}[\phi(Y)]$$
for all $a_i\geq 0$, $b\in(-\infty,\infty)$, and $m\geq 0$ (approximation of *increasing function* $\phi(\cdot)$ converges using simple functions from sequences $U_i$, $a_i$, and $b$ as $m\rightarrow\infty$). Therefore, $X\leq_{st}Y$ *if and only if* $\mathbb{E}[\phi(X)]\leq \mathbb{E}[\phi(Y)]$ holds for all increasing functions $\phi(\cdot)$ for which the expectation exists.

A more general definition can be stated, where two random variables $X$ and $Y$ satisfy $X\leq_{st}Y$ *if and only if* there exists two random variables $\hat{X}=_{st}X$ and $\hat{Y}=_{st}Y$ defined on the same probability space such that $\mathbb{P}(\hat{X}\leq\hat{Y})=1$.
