# Definitions
The set of all possible outcomes of a random experiment (cannot a priori know outcome) is known as the ***Sample Space***, $\Omega$, with each element within it called an *outcome*. A subset of *sample space* is called an ***event*** ($E\subseteq\Omega$); defined as ***mutually exclusive*** (disjoint) if $F\cap E=\emptyset$. Given a *sample space*, probability assigns number to *events* in $\Omega$ such that $0\leq \mathbb{P}(E)\leq 1$, $\mathbb{P}(\Omega)=1$, and for any sequence of *mutually exclusive events* $\mathbb{P}(\bigcup_{n=1}^{\infty}E_n)=\sum_{n=1}^{\infty}\mathbb{P}(E_n)$.

***Probability Space***, $(\Omega, \mathcal{F}, \mathbb{P})$; defines the *sample space* $\Omega$, ***event space*** defined as the set of all *events* $\mathcal{F}\subseteq\mathcal{P}(\Omega)$ (***power set*** of the set of all possible *outcomes*), and the ***probability function***/measure $\mathbb{P}$.

$\mathcal{F}$ is ***$\sigma$-algebra*** if $\Omega\in\mathcal{F}$, $A\in\mathcal{F}\rightarrow A^{\text{c}}\in\mathcal{F}$, and $A,B\in\mathcal{F}\rightarrow A\cup B\in\mathcal{F}$.
- **Example**) For $\Omega=\{1,2,3,4,5,6\}$, what would $\mathcal{F}=\{\Omega, \emptyset, \{1,2\}, \{3\}\}$ require to be $\sigma$-algebra. As $\{1,2\}\in\mathcal{F}$ then $\{1,2\}^{\text{c}}=\{3,4,5,6\}$ must also be in $\mathcal{F}$, similarly to $\{3\}\rightarrow\{1,2,4,5,6\}$. As $\{1,2\},\{3\}\in\mathcal{F}$ then so does their union, $\{1,2,3\}$, and again their complement $\{4,5,6\}$; this is results in $\mathcal{F}=\{\Omega, \emptyset, \{1,2\}, \{3,4,5,6\}, \{3\}, \{1,2,4,5,6\}, \{1,2,3\}, \{4,5,6\}\}$.
- **Example**) If $A,B\in\mathcal{F}$ then does $A\cap B\in\mathcal{F}$? $A,B\in\mathcal{F}\rightarrow A^{\text{c}},B^{\text{c}}\in\mathcal{F}\rightarrow A^{\text{c}}\cup B^{\text{c}}\in\mathcal{F}\rightarrow (A^{\text{c}}\cup B^{\text{c}})^{\text{c}}=A\cap B\in\mathcal{F}$, Yes.

The *probability function* $\mathbb{P}$ on $\mathcal{F}$ satisfies $\mathbb{P}(A)\geq0\;\forall A\in\mathcal{F}$, $\mathbb{P}(\Omega)=1$, and $A,B\in\mathcal{F} \land A\cap B=\emptyset\rightarrow \mathbb{P}(A\cup B)=\mathbb{P}(A)+\mathbb{P}(B)$
- **Example**) Find $\mathbb{P}(\emptyset)$ by the definition. As $\emptyset,\Omega\in\mathcal{F}$ and that $\Omega\land\emptyset=\emptyset$ then $\mathbb{P}(\emptyset\cup\Omega)=\mathbb{P}(\emptyset)+\mathbb{P}(\Omega)\rightarrow \mathbb{P}(\Omega)=\mathbb{P}(\emptyset)+\mathbb{P}(\Omega)$ and so $\mathbb{P}(\emptyset)=0$.
- **Example**) Does the behaviour $A,B\in\mathcal{F}\land A\subseteq B\rightarrow \mathbb{P}(B)\geq \mathbb{P}(A)$ exist with the given axioms. As $A,B\in\mathcal{F}$ then so do $A^{\text{c}},B^{\text{c}}\in\mathcal{F}$, and because $A\subseteq B$ it means that either $A^{\text{c}}\cap B=\emptyset$ (if $A=B$) or $A^{\text{c}}\cap B\neq\emptyset$ (if $A\subset B$) and so $\mathbb{P}(A^{\text{c}}\cap B)\geq 0$. Therefore, $\mathbb{P}(B)=\mathbb{P}(A)+\mathbb{P}(A^{\text{c}}\cap B)\geq \mathbb{P}(A)$.

Given $(\Omega, \mathcal{F}, \mathbb{P})$, *events* $A$ and $B$ are defined to be ***independent***, $A\mathrel{\perp\!\!\!\perp}B$, if $\mathbb{P}(A\cap B)=\mathbb{P}(A)\mathbb{P}(B)$. Further, where ***conditional probability*** is defined as $\mathbb{P}(A|B)=\frac{\mathbb{P}(A\cap B)}{\mathbb{P}(B)}$ (posterior ***Bayes Theorem*** $\mathbb{P}(B|A)=\frac{\mathbb{P}(A|B)\mathbb{P}(B)}{\mathbb{P}(A)}$); if the *events* are *independent*, then $\mathbb{P}(A|B)=\frac{\mathbb{P}(A)\mathbb{P}(B)}{\mathbb{P}(B)}=\mathbb{P}(A)$. Additionally, if $A\mathrel{\perp\!\!\!\perp}B$ then $\mathbb{P}(B)=\mathbb{P}(A\cap B)+\mathbb{P}(A^{\text{c}}\cap B)$ (disjoint probabilities) and so $\mathbb{P}(B)=\mathbb{P}(A)\mathbb{P}(B)+\mathbb{P}(A^{\text{c}}\cap B)\rightarrow \mathbb{P}(A^{\text{c}}\cap B)=\mathbb{P}(B)(1-\mathbb{P}(A))=\mathbb{P}(B)\mathbb{P}(A^{\text{c}})$; thus meaning that $A^{\text{c}}\mathrel{\perp\!\!\!\perp}B$. 

$E_1,\ldots,E_k$ are said to ***partition*** $\Omega$ if $E_i\cap E_j=\emptyset\;\forall i\neq j$ and $\bigcup_{n=1}^{k}E_n=\Omega$; this means that an *event*, $A$, can be re-defined as $A=\bigcup_{n=1}^{k}(E_n\cap A)\leftrightarrow \mathbb{P}(A)=\sum_{n=1}^{k}\mathbb{P}(E_n\cap A)=\sum_{n=1}^{k}\mathbb{P}(A|E_n)\mathbb{P}(E_n)$ based of the *partition* of the *sample space*.

Given a probability space, $(\Omega, \mathcal{F}, \mathbb{P})$, a ***random variable*** is a function $X:\Omega\rightarrow\mathbb{R}$ that is *$\mathcal{F}$ measurable*; meaning that $\forall c\in\mathbb{R}\;\{\omega\in\Omega| X(\omega)\leq c\}\in\mathcal{F}$.
- **Example**) Given $\Omega=\{1,2,3,4,5,6\}$, and $\mathcal{F}=\{\{1,3\},\{2,4,5,6\},\{1,2,3,4,5,6\},\emptyset\}$; is $\forall\omega\in\Omega\;X(\omega)=\omega$ *$\mathcal{F}$ measurable*? Consider $c=3$, as $\{\omega\in\Omega| X(\omega)\leq 3\}=\{1,2,3\}\not\in\mathcal{F}$ and so it is not *$\mathcal{F}$ measurable*.

$B_1\subseteq B_2\subseteq\ldots=\{B_n\}_{n\geq1}\rightarrow \lim_{n\rightarrow\infty}B_n=\bigcup_{i=1}^{\infty}B_i$ can then be thought of as an analogue as a monotonically increasing sequence $x$ converging at $x_0$ (further $\lim_{n\rightarrow\infty}\mathbb{P}(B_n)=\mathbb{P}(\bigcup_{j=1}^{\infty}B_j)$); and $B_1\supseteq B_2\ldots=\{B_n\}_{n\geq1}\rightarrow \lim_{n\rightarrow\infty}B_n=\bigcap_{j=1}^{\infty}B_j$ can then be thought of as an analogue as a monotonically decreasing sequence $y$ converging at $y_0$ (further $\lim_{n\rightarrow\infty}\mathbb{P}(B_n)=\mathbb{P}(\bigcap_{j=1}^{\infty}B_j)$).

Given a *random variable* $X$ on $(\Omega, \mathcal{F}, \mathbb{P})$, then the ***cumulative distribution function*** (*cdf*) is defined as $\mathbb{P}(\{\omega\in\Omega| X(\omega)\leq c\})=\mathbb{P}(X\leq c)=F_X(c)$ (follows that $\mathbb{P}(X<c)=F_X(c^-)$ and $\mathbb{P}(X=c)=F_X(c)-F_X(c^-)$).

A function $F$ is the *cdf* of some *random variable* if the following properties hold:
1. $F$ is a non-decreasing. Given $x\geq y$, then $F(x)=\mathbb{P}(\{\omega\in\Omega| X(\omega)\leq x\})=\mathbb{P}(\{\omega\in\Omega| X(\omega)\leq y\})+\mathbb{P}(\{\omega\in\Omega| y<X(\omega)\leq x\})=F(y)+\mathbb{P}(y<X\leq x)$, and therefore $F(x)\geq F(y)$; meaning that $F$ is non-decreasing.
2. $\lim_{x\rightarrow\infty}F(x)=1$ and $\lim_{x\rightarrow-\infty}F(x)=0$. As $\lim_{x\rightarrow\infty}F(x)=\mathbb{P}(\bigcup_{j=1}^{\infty}B_j)=\mathbb{P}(\Omega)=1$ then $\forall\epsilon >0$, $\exists N_\epsilon$ such that $\forall n\geq N_\epsilon$ $F(n)\geq 1-\epsilon$; similarly, as $\lim_{x\rightarrow-\infty}F(x)=\mathbb{P}(\bigcap_{j=1}^{\infty}B_j)=\mathbb{P}(\emptyset)=0$.
3. $F$ is a *right continuous* function ($\lim_{x\rightarrow c^+}F(x)=F(c)$).
