A ***Group***, $(G,*)$, is a *set* $G$, coupled with a *binary operator* (closed operator $a,b\in G\rightarrow a*b\in G$ which is *'well defined'*) $*$ (for example $(\mathbb{R},+)$ is the real numbers with addition) such that $*$ is associative ($a,b,c\in G\rightarrow (a*b)*c=a*(b*c)$), neutral element $e$ of $G$ exists ($\exists e\in G$ such that $\forall a\in G$, $e*a=a*e=a$), and that every element in $G$ has an *inverse* ($\forall a\in G,\;\exists a^{-1}\in G$ such that $a*a^{-1}=a^{-1}*a=e$). If there exists *commutativity* ($a,b\in G\rightarrow a*b=b*a$) in the group it is called an *abelian group*.

A ***Ring***, $(R,+,\times)$, is a *set* $R$ with two binary operations $+$ and $\times$ such that; $(R,+)$ is an *abelian group* where *neutral* $e=0$; and for $(R,\times)$ $\times$ is *associative*, $i\in R$ is *neutral*, and $\forall a,b,c\in R\; a\times(b+c)=a\times b+a\times c$ and $(b+c)\times a=b\times a+c\times a$ ($\times$ distributes over $+$).
- **Example**) Is $(\mathbb{Z}[\frac{1+\sqrt{5}}{2}],+,\times)$ a ring where $\mathbb{Z}[\omega]=\{a_0+a_1\omega^1+\ldots+a_n\omega^n : a_j\in\mathbb{Z}\land n\in\mathbb{Z}^+\}$? If $\omega=\frac{1+\sqrt{5}}{2}$ then $\omega^2=1+\omega$ and so $\mathbb{Z}[\frac{1+\sqrt{5}}{2}]=\{a+b\omega:a,b\in\mathbb{Z}\}\subset \mathbb{R}$. Therefore, associativity (commutative) is inherited from $(\mathbb{R},+,\times)$. Additionally, the additive neutral element is $e=0$ as if $a\in\mathbb{Z}[\frac{1+\sqrt{5}}{2}]\rightarrow a+0=0+a=a$; and if $a=a_1+a_2\omega\in\mathbb{Z}[\frac{1+\sqrt{5}}{2}]$, then $a^{-1}=-a=-(a_1+a_2\omega)\in\mathbb{Z}[\frac{1+\sqrt{5}}{2}]$ is the additive inverse due to $a+a^{-1}=a^{-1}+a=0=e$. For $a,b,c\in\mathbb{Z}[\frac{1+\sqrt{5}}{2}]$ $a\times(b+c)=(a_1+a_2\omega)([b_1+c_1]+[b_2+c_2]\omega)=[a_1b_1+a_1c_1+a_2b_2+a_2c_2]+[a_1b_2+a_1c_2+a_2b_1+a_2c_1+a_2b_2+a_2c_2]\omega=([a_1b_1+a_2b_2]+[a_1b_2+a_2b_1+a_2b_2]\omega)+([a_1c_1+a_2c_2]+[a_1c_2+a_2c_1+a_2c_2]\omega)=a\times b+a\times c$. Therefore, $(\mathbb{Z}[\frac{1+\sqrt{5}}{2}],+,\times)$ is indeed a ring.

# Unit 1&2: Basics with Julia
*Julia*, https://julialang.org/, is a modern, high-performance, and dynamic language designed for numerical and scientific computing; combining the ease of use of scripting languages like Python, the speed of compiled languages like C or Fortran, and the optimised matrix and statistical operations of MATLAB and R.
- Based on the *LLVM* infrastructure; a set of compiler and toolchain technologies that can be used to develop a frontend for any programming language and a backend for any instruction set architecture. Thus, allowing for language-independent intermediate representation that serves as a portable, high-level assembly language that can be optimized with a variety of transformations over multiple passes.

**Linear** algebra is defined by vectors/matrices `A=[a_11 a_12 ... a_1n; a_21 a_22 ... a_2n; ...; a_m1 a_m2 ... a_mn]` called arrays (note that indexing begins with $1$ ends with $n$; indexing from the end is done via `end-x`), they have the specialised functions `dot(x, y)` (dot product), `det(A)` (determinant), `tr(A)` (trace), `eigvals(A)` (eigenvalues), and `eigvecs(A)` (eigenvectors); additioanlly, for every binary operator there is a corresponding "dot" operation `.*`, ect. which will automatically perform element-wise operation on teh arrays (vectors/matrices).

Functions, loops, and if statements can be constructed as seen below (algorithm to calculate fibbonacci sequence with $O(n\log n)$ time and $O(1)$ space, given that power calculation take $O(\log n)$ times. Uses the idea that $\phi^2=1+\phi$ to construct - can be refined further to $O(\log n)$ times for $F_n=\frac{\phi^n-(-\phi)^{-n}}{2\phi-1}$)

```julia
function fibbonacci_sum(n::Integer)
    if n < 1
        return 1
    end
     = big((1+(5))/(2))
    fib = 0
    for i in 1:n
        fib += big((-1)^(i+1)*^(n+1-2*i))
    end
    return fib
end
```

The *gamma* function, $\Gamma(z) = \int_0^\infty x^{z-1}e^{-x}dx$, has the unique behaviour of $\Gamma(z+1)=\int_0^\infty x^{z}e^{-x}dx=[-x^{z}e^{-x}]_0^\infty+z\int_0^\infty x^{z-1}e^{-x}dx=z\Gamma(z)=\ldots=z!\Gamma(0)=z!$ and so $n!=\Gamma(n+1)$; and so can be thought o fhte continuous analogue to the factorial function. Implementing this can be done via the `SpecialFunctions` library or estimated via *Riemann Sum* approximation:

```julia
using SpecialFunctions
real = gamma(6.5)
function gamma_estimate(z::Number, delta::Number=0.01, M::Number=50)
    integrand(x) = x^(z-1)*exp(-x)
    x_grid = 0:delta:M
    return sum(delta*integrand(x) for x in x_grid)
end
estimate = gamma_estimate(6.5)
println(real - estimate)
```

```julia
# O(n^2) time O(1) space
function bubble_sort!(list::Vector{<:Number})
    n = length(list)
    for i in 1:n
        for j in 1:n-i
            if list[j]>list[j+1]
                list[j],list[j+1] = list[j+1],list[j]
            end
        end
    end
    return list
end

# O(nlogn) time O(1) space
function merge_sort!(list::Vector{<:Number}, verbose::Bool=false, start::Integer=1, stop::Integer=length(list), level::Integer=0)
    verbose && println("   "^level, list[start:stop])
    if stop-start < 1
        return list
    end
    halfway = floor(Integer, (stop-start)/2)
    left = merge_sort!(list, verbose, start, start+halfway, level+1)
    right = merge_sort!(list, verbose, start+halfway+1, stop, level+1)
    i = start
    j = start+halfway+1
    while j  stop && i < j
        if list[i]  list[j]
            i += 1
        else
            list[i+1:j], list[i] = list[i:j-1], list[j]
            i += 1
            j += 1
        end
    end
    verbose && println("   "^level, list[start:stop])
    return list
end

# O(nlogn) time O(1) space
function quick_sort!(list::Vector{<:Number}, verbose::Bool=false, start::Integer=1, stop::Integer=length(list), level::Integer=0)
    verbose && println("   "^level, list[start:stop])
    if stop-start < 1
        return list
    end
    min_value = minimum(list[start:stop])
    max_value = maximum(list[start:stop])
    pivot = min_value + (max_value-min_value)/2
    i = start
    j = stop
    while j > i
        if list[i]  pivot && list[j] < pivot
            list[i], list[j] = list[j], list[i]
        end
        if list[i] < pivot
            i += 1
        end
        if list[j]  pivot
            j -= 1
        end
    end
    if j==start
        return list
    end
    j = list[j]  pivot ? j-1 : j
    left = quick_sort!(list, verbose, start, j, level+1)
    right = quick_sort!(list, verbose, j+1, stop, level+1)
    verbose && println("   "^level, list[start:stop])
    return list
end

```

```julia
using Random
Random.seed!(0)

data = rand(Int, 10^5)

print("Bubble Sort: ")
@time bubble_sorted = bubble_sort!(data)
println("sorted=$(issorted(bubble_sorted))")

print("Merge Sort: ")
@time merge_sorted = merge_sort!(data)
println("sorted=$(issorted(merge_sorted))")

print("Quick Sort: ")
@time quick_sorted = quick_sort!(data)
println("sorted=$(issorted(quick_sorted))")
```
