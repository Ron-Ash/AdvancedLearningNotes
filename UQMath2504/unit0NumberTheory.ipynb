A ***Group***, $(G,*)$, is a *set* $G$, coupled with a *binary operator* (closed operator $a,b\in G\rightarrow a*b\in G$ which is *'well defined'*) $*$ (for example $(\mathbb{R},+)$ is the real numbers with addition) such that $*$ is associative ($a,b,c\in G\rightarrow (a*b)*c=a*(b*c)$), neutral element $e$ of $G$ exists ($\exists e\in G$ such that $\forall a\in G$, $e*a=a*e=a$), and that every element in $G$ has an *inverse* ($\forall a\in G,\;\exists a^{-1}\in G$ such that $a*a^{-1}=a^{-1}*a=e$). If there exists *commutativity* ($a,b\in G\rightarrow a*b=b*a$) in the group it is called an *abelian group*.

A ***Ring***, $(R,+,\times)$, is a *set* $R$ with two binary operations $+$ and $\times$ such that; $(R,+)$ is an *abelian group* where *neutral* $e=0$; and for $(R,\times)$ $\times$ is *associative*, $i\in R$ is *neutral*, and $\forall a,b,c\in R\; a\times(b+c)=a\times b+a\times c$ and $(b+c)\times a=b\times a+c\times a$ ($\times$ distributes over $+$).
- **Example**) Is $(\mathbb{Z}[\frac{1+\sqrt{5}}{2}],+,\times)$ a ring where $\mathbb{Z}[\omega]=\{a_0+a_1\omega^1+\ldots+a_n\omega^n : a_j\in\mathbb{Z}\land n\in\mathbb{Z}^+\}$? If $\omega=\frac{1+\sqrt{5}}{2}$ then $\omega^2=1+\omega$ and so $\mathbb{Z}[\frac{1+\sqrt{5}}{2}]=\{a+b\omega:a,b\in\mathbb{Z}\}\subset \mathbb{R}$. Therefore, associativity (commutative) is inherited from $(\mathbb{R},+,\times)$. Additionally, the additive neutral element is $e=0$ as if $a\in\mathbb{Z}[\frac{1+\sqrt{5}}{2}]\rightarrow a+0=0+a=a$; and if $a=a_1+a_2\omega\in\mathbb{Z}[\frac{1+\sqrt{5}}{2}]$, then $a^{-1}=-a=-(a_1+a_2\omega)\in\mathbb{Z}[\frac{1+\sqrt{5}}{2}]$ is the additive inverse due to $a+a^{-1}=a^{-1}+a=0=e$. For $a,b,c\in\mathbb{Z}[\frac{1+\sqrt{5}}{2}]$
 $a\times(b+c)=(a_1+a_2\omega)([b_1+c_1]+[b_2+c_2]\omega)=[a_1b_1+a_1c_1+a_2b_2+a_2c_2]+[a_1b_2+a_1c_2+a_2b_1+a_2c_1+a_2b_2+a_2c_2]\omega=([a_1b_1+a_2b_2]+[a_1b_2+a_2b_1+a_2b_2]\omega)+([a_1c_1+a_2c_2]+[a_1c_2+a_2c_1+a_2c_2]\omega)=a\times b+a\times c$. Therefore, $(\mathbb{Z}[\frac{1+\sqrt{5}}{2}],+,\times)$ is indeed a ring.
